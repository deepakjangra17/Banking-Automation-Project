```python

try:
    conobj=sqlite3.connect(database="Banking.sqlite")
    curobj=conobj.cursor()
    curobj.execute("create table accounts(acn integer primary key autoincrement,name text,pass text,email text,mob text,bal float,type text,opendate text)")
    curobj.execute("create table txns(acn int,amt float,updatebal float,type text,txndate text)")
    conobj.close()
    print("tables created")
except:
    print("something went wrong")


win=Tk()
win.state("zoomed")
win.configure(bg="powder blue")
win.resizable(width=False,height=False)

def frm1():
    frm2=Frame(win)
    frm2=Frame(win,highlightbackground='black',highlightthickness=2)
    frm2.place(relx=0,rely=0,relwidth=1,relheight=.15)
    frm2.configure(bg="powder blue",bd=25)
    

    title=Label(win,text="Banking Automation",font=("arial",55,"bold","underline"),bg="powder blue")
    title.place(relx=0.24,rely=0.01)


    date=Label(win,text=f"{datetime.now().date()}",font=("arial",10,"bold"),bg="powder blue")
    date.place(relx=0.91,rely=0.1)

def mainscreen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="lavender")

    def new():
        frm.destroy()
        newscreen()
    
    def fp():
        frm.destroy()
        fpscreen()
        
    def reset():
        e_acn.delete(0,"end")
        e_pass.delete(0,"end")
        e_acn.focus()

    def login():
        acn=e_acn.get()
        pwd=e_pass.get()
        
        if(len(acn)==0 or len(pwd)==0):
            messagebox.showerror("Login","Empty fields are not allowed!")
        else:
            conobj=sqlite3.connect(database="Banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select * from accounts where acn=? and pass=?",(acn,pwd))
            tup=curobj.fetchone()
            if tup==None:
                messagebox.showerror("Login","Invalid ACN/PASS")
            else:
                global uacn,uname
                uacn=tup[0]
                uname=tup[1]
                frm.destroy()
                homescreen()
        
        
    lbl_acn=Label(frm,text="Account no.",font=('arial',20,"bold"),bg="lavender")
    lbl_acn.place(relx=0.32,rely=0.2)
    
    e_acn=Entry(frm,font=("arial",20,"bold"),bd=5)
    e_acn.place(relx=0.45,rely=0.2)
    e_acn.focus()
    
    lbl_pass=Label(frm,text="Password",font=('arial',20,"bold"),bg="lavender")
    lbl_pass.place(relx=0.32,rely=0.3)
    
    e_pass=Entry(frm,font=("arial",20,"bold"),bd=5,show="*")
    e_pass.place(relx=0.45,rely=0.3)
       
    lgn_btn=Button(frm,command=login,text="Login",font=('arial',15,"bold"),bd=5)
    lgn_btn.place(relx=0.49,rely=0.45)

    reset_btn=Button(frm,command=reset,text="Reset",font=('arial',15,"bold"),bd=5)
    reset_btn.place(relx=0.575,rely=0.45)
    
    fp_btn=Button(frm,command=fp,text="Forgot password",width=15,font=('arial',15,"bold"),bd=5)
    fp_btn.place(relx=0.49,rely=0.54)
    
    lbl_nu=Label(frm,text="new user",font=('arial',15,"italic"),bg="lavender")
    lbl_nu.place(relx=0.43,rely=0.64)
    
    cu_btn=Button(frm,text="Signup",font=('arial',15,"bold"),bd=5,command=new)
    cu_btn.place(relx=0.53,rely=0.63)
    
    
    
def newscreen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="lavender")
    
    def back():
        frm.destroy()
        mainscreen()
        
    def reset():
        e_name.delete(0,"end")
        e_pass.delete(0,"end")
        e_email.delete(0,"end")
        e_mob.delete(0,"end")
        e_name.focus()
    
    def openacndb():
        bal=0
        opendate=time.ctime()
        name=e_name.get()
        pwd=e_pass.get()
        email=e_email.get()
        mob=e_mob.get()
        acntype=cb_type.get()
        
        if(len(name)==0 or len(pwd)==0 or len(email)==0 or len(mob)==0 or len(acntype)==0):
            messagebox.showerror("Login","Empty fields are not allowed!")
        else:
        
            conobj=sqlite3.connect(database="Banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("insert into accounts(name,pass,email,mob,bal,type,opendate) values(?,?,?,?,?,?,?)",(name,pwd,email,mob,bal,acntype,opendate))
            conobj.commit()
            curobj.close()
            curobj=conobj.cursor()
            curobj.execute("select max(acn) from accounts")
            tup=curobj.fetchone()
            messagebox.showinfo("New Account",f"Account opened with ACN :{tup[0]}")
            conobj.close()
        
    
    lbl_name=Label(frm,text="Name",font=('arial',20,"bold"),bg="lavender")
    lbl_name.place(relx=0.3,rely=0.2)
    
    e_name=Entry(frm,font=("arial",20,"bold"),bd=5)
    e_name.place(relx=0.45,rely=0.2)
    e_name.focus()
    
    lbl_pass=Label(frm,text="Password",font=('arial',20,"bold"),bg="lavender")
    lbl_pass.place(relx=0.3,rely=0.3)
    
    e_pass=Entry(frm,font=("arial",20,"bold"),bd=5)
    e_pass.place(relx=0.45,rely=0.3)
    
    lbl_email=Label(frm,text="Email",font=('arial',20,"bold"),bg="lavender")
    lbl_email.place(relx=0.3,rely=0.4)
    
    e_email=Entry(frm,font=("arial",20,"bold"),bd=5)
    e_email.place(relx=0.45,rely=0.4)
    
    lbl_mob=Label(frm,text="Mobile No.",font=('arial',20,"bold"),bg="lavender")
    lbl_mob.place(relx=0.3,rely=0.5)
    
    e_mob=Entry(frm,font=("arial",20,"bold"),bd=5)
    e_mob.place(relx=0.45,rely=0.5)
    
        
    lbl_type=Label(frm,text="Account type",font=('arial',20,"bold"),bg="lavender")
    lbl_type.place(relx=0.3,rely=0.6)
    
    cb_type=Combobox(frm,values=['Saving','Current'],font=('arial',20,'bold'))
    cb_type.current(0)
    cb_type.place(relx=0.45,rely=0.6)
    
    create_btn=Button(frm,command=openacndb,text="Create",font=('arial',15,"bold"),bd=5)
    create_btn.place(relx=0.5,rely=0.75)

    reset_btn=Button(frm,command=reset,text="Reset",font=('arial',15,"bold"),bd=5)
    reset_btn.place(relx=0.58,rely=0.75)

    back_btn=Button(frm,text="Back",font=('arial',15,"bold"),bd=5,command=back)
    back_btn.place(relx=0,rely=0)

    
def fpscreen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="lavender")
    
    def back():
        frm.destroy()
        mainscreen()
        
    def reset():
        e_acn.delete(0,"end")
        e_email.delete(0,"end")
        e_mob.delete(0,"end")
        e_acn.focus()

    def recpassdb():
        acn=e_acn.get()
        email=e_email.get()
        mob=e_mob.get()
        
        conobj=sqlite3.connect(database="Banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select pass from accounts where acn=? and email=? and mob=?",(acn,email,mob))
        tup=curobj.fetchone()
        if tup==None:
            messagebox.showerror("Forgot Password","Account does not exist")
        else:
            messagebox.showinfo("Forgot Password",f"Your password : {tup[0]}")
        
        
        
    back_btn=Button(frm,text="Back",font=('arial',15,"bold"),bd=5,command=back)
    back_btn.place(relx=0,rely=0)


    lbl_acn=Label(frm,text="Account No.",font=('arial',20,"bold"),bg="lavender")
    lbl_acn.place(relx=0.32,rely=0.2)
    
    e_acn=Entry(frm,font=("arial",20,"bold"),bd=5)
    e_acn.place(relx=0.45,rely=0.2)
    e_acn.focus()
 
    
    lbl_email=Label(frm,text="Email",font=('arial',20,"bold"),bg="lavender")
    lbl_email.place(relx=0.32,rely=0.3)
    
    e_email=Entry(frm,font=("arial",20,"bold"),bd=5)
    e_email.place(relx=0.45,rely=0.3)
    
    lbl_mob=Label(frm,text="Mobile No.",font=('arial',20,"bold"),bg="lavender")
    lbl_mob.place(relx=0.32,rely=0.4)
    
    e_mob=Entry(frm,font=("arial",20,"bold"),bd=5)
    e_mob.place(relx=0.45,rely=0.4)
    
    
    rec_btn=Button(frm,command=recpassdb,text="Recover",font=('arial',15,"bold"),bd=5)
    rec_btn.place(relx=0.49,rely=0.55)

    reset_btn=Button(frm,command=reset,text="Reset",font=('arial',15,"bold"),bd=5)
    reset_btn.place(relx=0.575,rely=0.55)

    
def homescreen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="lavender")
    
    def logout():
        frm.destroy
        mainscreen()
    
    def detailsscreen():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.place(relx=.25,rely=.15,relwidth=.6,relheight=.6)
        ifrm.configure(bg="white")
        
        lbl=Label(ifrm,text="This is Account Details Screen",font=('arial',25,'bold'),bg="white")
        lbl.pack()
        
        conobj=sqlite3.connect(database="Banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select * from accounts where acn=?",(uacn,))
        tup=curobj.fetchone()
        
        lbl_acn=Label(ifrm,text=f"Account No.\t : {tup[0]}",font=('arial',15),bg="white")
        lbl_acn.place(relx=0.2,rely=0.2)
        
        lbl_bal=Label(ifrm,text=f"Account Balance\t : {tup[5]}",font=('arial',15),bg="white")
        lbl_bal.place(relx=0.2,rely=0.3)
        
        lbl_open=Label(ifrm,text=f"Open date\t : {tup[7]}",font=('arial',15),bg="white")
        lbl_open.place(relx=0.2,rely=0.4)
        
    
    def depositscreen():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.place(relx=.25,rely=.15,relwidth=.6,relheight=.6)
        ifrm.configure(bg="white")
        
        def depdb():
            amt=float(e_amt.get())
            conobj=sqlite3.connect(database="Banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select bal from accounts where acn=?",(uacn,))
            bal=curobj.fetchone()[0]
            curobj.close()
            
            curobj=conobj.cursor()
            curobj.execute("update accounts set bal=bal+? where acn=?",(amt,uacn))
            curobj.execute("insert into txns values(?,?,?,?,?)",(uacn,amt,bal+amt,"Cr.",time.ctime()))
            conobj.commit()
            conobj.close()
            
            messagebox.showinfo("Amount deposited",f"Updated balance: {bal+amt}")
            
            
            
            
        lbl=Label(ifrm,text="This is Deposit Screen",font=('arial',25,'bold'),bg="white")
        lbl.pack()
        
        lbl_amt=Label(ifrm,text="Amount",font=('arial',15),bg="white")
        lbl_amt.place(relx=0.2,rely=0.2)
        
        e_amt=Entry(ifrm,font=('arial',15,'bold'),bd=5)
        e_amt.place(relx=.38,rely=0.2)
        
        dep_btn=Button(ifrm,command=depdb,text='Deposit',font=('arial',15,'bold'),bd=5)
        dep_btn.place(relx=.4,rely=.4)
        
    def withdrawscreen():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.place(relx=.25,rely=.15,relwidth=.6,relheight=.6)
        ifrm.configure(bg="white")
        
        def withdrawdb():
            amt=float(e_amt.get())
            conobj=sqlite3.connect(database="Banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select bal from accounts where acn=?",(uacn,))
            bal=curobj.fetchone()[0]
            curobj.close()
            if bal>amt:
                curobj=conobj.cursor()
                curobj.execute("update accounts set bal=bal-? where acn=?",(amt,uacn))
                curobj.execute("insert into txns values(?,?,?,?,?)",(uacn,amt,bal+amt,"Db.",time.ctime()))
                conobj.commit()
                messagebox.showinfo("Amount withdraw",f"Updated balance: {bal-amt}")

            else:
                messagebox.showwarning("Withdraw",f"Insufficient Balance:{bal}")
            conobj.close()
            
                       
        lbl=Label(ifrm,text="This is Withdraw Screen",font=('arial',25,'bold'),bg="white")
        lbl.pack()
        
        lbl_amt=Label(ifrm,text="Amount",font=('arial',15),bg="white")
        lbl_amt.place(relx=0.2,rely=0.2)
        
        e_amt=Entry(ifrm,font=('arial',15,'bold'),bd=5)
        e_amt.place(relx=.38,rely=0.2)
        
        withdraw_btn=Button(ifrm,command=withdrawdb,text='Withdraw',font=('arial',15,'bold'),bd=5)
        withdraw_btn.place(relx=.4,rely=.4)
        
    def historyscreen():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.place(relx=.25,rely=.15,relwidth=.6,relheight=.6)
        ifrm.configure(bg="white")
        
        lbl=Label(ifrm,text="This is History Screen",font=('arial',25,'bold'),bg="white")
        lbl.pack()
        
        tv=Treeview(ifrm)
        tv.place(x=60,y=100,height=100,width=600)
        
        
        style=Style()
        style.configure("Treeview.Heading", font=('Arial',10,'bold'),foreground='brown')

        sb=Scrollbar(ifrm,orient='vertical',command=tv.yview)
        sb.place(x=650,y=100,height=100)
        tv.configure(yscrollcommand=sb.set)
        
        tv['columns']=('date','amt','type','upbal')
        
        tv.column('date',width=80,anchor='c')
        tv.column('amt',width=40,anchor='c')
        tv.column('type',width=40,anchor='c')
        tv.column('upbal',width=40,anchor='c')
        
        tv.heading('date',text='Txn Date')
        tv.heading('amt',text='Amount')
        tv.heading('type',text='Txn type')
        tv.heading('upbal',text='Updated Bal')
        
        tv['show']='headings'
        
        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select * from txns where acn=?",(uacn,))
        
        for row in curobj:
            tv.insert("","end",values=(row[4],row[1],row[3],row[2]))
        
        
        
    def transferscreen():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.place(relx=.25,rely=.15,relwidth=.6,relheight=.6)
        ifrm.configure(bg="white")
        
        lbl=Label(ifrm,text="This is Transfer Screen",font=('arial',25,'bold'),bg="white")
        lbl.pack()
        
        def transfer():
            to=e_to.get()
            amt=float(e_amt.get())
            
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select * from accounts where acn=?",(to,))
            tup=curobj.fetchone()
            curobj.close()
            if tup==None:
                messagebox.showerror("Transfer","To ACN does not exist")
            else:
                curobj=conobj.cursor()
                curobj.execute("select bal from accounts where acn=?",(uacn,))
                bal=curobj.fetchone()[0]
                curobj.close()
                if bal>=amt:
                    curobj=conobj.cursor()
                    curobj.execute("update accounts set bal=bal+? where acn=?",(amt,to))
                    curobj.execute("update accounts set bal=bal-? where acn=?",(amt,uacn))
                    curobj.execute("insert into txns values(?,?,?,?,?)",(to,amt,bal+amt,"Cr.",time.ctime()))
                    curobj.execute("insert into txns values(?,?,?,?,?)",(uacn,amt,bal-amt,"Db.",time.ctime()))
                    conobj.commit()
                    conobj.close()
                    messagebox.showinfo("Transfer",f"Rs. {amt} is transfered to {to} ACN")
                    
                else:
                    messagebox.showinfo("Transfer",f"Insufficient Bal:{bal}")
                
        lbl_to=Label(ifrm,text="To Account",font=('arial',15,'bold'),bg="white")
        lbl_to.place(relx=.2,rely=.3)
        
        e_to=Entry(ifrm,font=('arial',15,'bold'),bd=5)
        e_to.place(relx=.35,rely=.3)
        
        lbl_amt=Label(ifrm,text="Amount",font=('arial',15,'bold'),bg="white")
        lbl_amt.place(relx=.2,rely=.4)
        
        e_amt=Entry(ifrm,font=('arial',15,'bold'),bd=5)
        e_amt.place(relx=.35,rely=.4)
        
        btn=Button(ifrm,command=transfer,text="Transfer",font=('arial',15,'bold'),bd=5)
        btn.place(relx=.4,rely=.5)
    
    def updateprofile():
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.place(relx=.25,rely=.15,relwidth=.6,relheight=.6)
        ifrm.configure(bg="white")
        
        lbl=Label(ifrm,text="This is Update Profile Screen",font=('arial',25,'bold'),bg="white")
        lbl.pack()
        
        def update():
            name=e_name.get()
            pwd=e_pass.get()
            email=e_email.get()
            mob=e_mob.get()
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("update accounts set name=?,pass=?,email=?,mob=? where acn=?",(name,pwd,email,mob,uacn))
            conobj.commit()
            conobj.close()
            messagebox.showinfo("Update","Profile Updated")
            ifrm.destroy()
            global uname
            uname=name
            updateprofile()
        
    
        lbl_name=Label(ifrm,text="Name",font=('arial',15,'bold'),bg="white")
        lbl_name.place(relx=.1,rely=.2)
        
        e_name=Entry(ifrm,font=('arial',15,'bold'),bd=5)
        e_name.place(relx=.1,rely=.3)
        
        lbl_pass=Label(ifrm,text="Pass",font=('arial',15,'bold'),bg="white")
        lbl_pass.place(relx=.1,rely=.4)
        
        e_pass=Entry(ifrm,font=('arial',15,'bold'),bd=5)
        e_pass.place(relx=.1,rely=.5)
        
        lbl_email=Label(ifrm,text="Email",font=('arial',15,'bold'),bg="white")
        lbl_email.place(relx=.5,rely=.2)
        
        e_email=Entry(ifrm,font=('arial',15,'bold'),bd=5)
        e_email.place(relx=.5,rely=.3)
        
        lbl_mob=Label(ifrm,text="Mob",font=('arial',15,'bold'),bg="white")
        lbl_mob.place(relx=.5,rely=.4)
        
        e_mob=Entry(ifrm,font=('arial',15,'bold'),bd=5)
        e_mob.place(relx=.5,rely=.5)
        
        btn=Button(ifrm,text="update",font=('arial',15,'bold'),command=update)
        btn.place(relx=.5,rely=.7)
    
        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select * from accounts where acn=?",(uacn,))
        tup=curobj.fetchone()
        curobj.close()
        
        e_name.insert(0,tup[1])
        e_pass.insert(0,tup[2])
        e_email.insert(0,tup[3])
        e_mob.insert(0,tup[4])
    
       
    lbl_wel=Label(frm,text=f"Welcome,{uname}",font=('arial',15,'bold'),bg="lavender")
    lbl_wel.place(relx=0,rely=0)
    
    logout_btn=Button(frm,text="logout",font=('arial',15,'bold'),command=logout)
    logout_btn.place(relx=.93,rely=0)
    
    details_btn=Button(frm,text="Details",font=('arial',20,'bold'),bd=5,width=12,command=detailsscreen)
    details_btn.place(relx=0.02,rely=.1)
    
    deposit_btn=Button(frm,command=depositscreen,text="Deposit",font=('arial',20,'bold'),width=12,bd=5)
    deposit_btn.place(relx=0.02,rely=.2)
    
    withdraw_btn=Button(frm,command=withdrawscreen,text="Withdraw",font=('arial',20,'bold'),width=12,bd=5)
    withdraw_btn.place(relx=0.02,rely=.3)
    
    profile_btn=Button(frm,command=updateprofile,text="Update profile",font=('arial',20,'bold'),width=12,bd=5)
    profile_btn.place(relx=0.02,rely=.4)
    
    history_btn=Button(frm,command=historyscreen,text="History",font=('arial',20,'bold'),width=12,bd=5)
    history_btn.place(relx=0.02,rely=.5)
    
    trans_btn=Button(frm,command=transferscreen,text="Transfer",font=('arial',20,'bold'),width=12,bd=5)
    trans_btn.place(relx=0.02,rely=.6)
frm1()        
mainscreen()   
win.mainloop()
```
